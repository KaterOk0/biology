```python
from skimage import io as skio
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = skio.imread('image.png', as_grey=True)
plt.imshow(image)
```

```python
image = (image * 256).astype(int)
image[image == 256] = 255
plt.imshow(image)
```

```python
def defineMatrix(img, bins):
    in_bin = 256 // bins
    resultMatrix = np.zeros((bins, bins))
    img = img.copy()
    img //= in_bin
    print(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if i + 1 < img.shape[0]:
                resultMatrix[img[i, j], img[i + 1, j]] += 1
            if j + 1 < img.shape[1]:
                if i + 1 < img.shape[0]:
                    resultMatrix[img[i, j], img[i + 1, j + 1]] += 1
                resultMatrix[img[i, j], img[i, j + 1]] += 1
                if i > 0:
                    resultMatrix[img[i, j], img[i - 1, j + 1]] += 1
    return resultMatrix
```

```python
plt.imshow(defineMatrix(image, 16))
```

```python
plt.imshow(defineMatrix(image, 256))
```
