```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
img = ndimage.imread('hedgehog.jpg', mode='RGB')
plt.imshow(img)
```

```python
color = ('b', 'g', 'r')
for channel, col in enumerate(color):
    hist, bins = np.histogram(img[:, :, channel], 256, [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])
```
