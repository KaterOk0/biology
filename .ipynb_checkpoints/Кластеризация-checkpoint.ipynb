```python
from Bio import Phylo
tree = Phylo.read("treeT2.txt", "newick")
```

```python
import itertools
distance_matrix = {}
# перебираем все комбинации
list = tree.get_terminals()
for x,y in itertools.combinations(list, 2):
    v = tree.distance(x, y)
    distance_matrix[x.name] = distance_matrix.get(x.name, {})
    distance_matrix[x.name][y.name] = v
    distance_matrix[y.name] = distance_matrix.get(y.name, {})
    distance_matrix[y.name][x.name] = v
for x in tree.get_terminals():
    distance_matrix[x.name][x.name] = 0
```

```python
import pandas as pd
```

```python
dataFrame = pd.DataFrame(distance_matrix)
```

```python
dataFrame
```

```python
import numpy as np
dist_matrix = np.array(dataFrame.as_matrix())
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit(dataFrame)
centroids = kmeans.cluster_centers_
# fit & predict clusters
data = kmeans.fit_predict(dist_matrix)
```

```python
data.size
```

```python
cluster0 = []
cluster1 = []
cluster2 = []
for i in range(data.size):
    if data[i] == 0:
        cluster0.append(list[i])
    if data[i] == 1:
        cluster1.append(list[i])
    if data[i] == 2:
        cluster2.append(list[i])
```

```python
print("1: ")
for clade in cluster0:
    print(clade)
```

```python
print("2: ")
for clade in cluster1:
    print(clade)
```

```python
print("3: ")
for clade in cluster2:
    print(clade)
```
